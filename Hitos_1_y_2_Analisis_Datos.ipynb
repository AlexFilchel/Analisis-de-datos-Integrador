## Hito 1: Elección y Planteo

**Dataset elegido:** `student_learning_trajectory_noisy.csv`  

Este conjunto de datos simula el comportamiento de aprendizaje y el rendimiento de los estudiantes durante un semestre (16 semanas). Cada fila representa a un estudiante en una semana, capturando sus hábitos de estudio, factores de estilo de vida y resultados académicos. En esta entrega se trabaja con la versión `noisy`, ideal para justificar el proceso de limpieza y preparación de datos.

Es sintético, lo que significa que los datos se generan artificialmente utilizando reglas realistas y variaciones aleatorias. No contiene información real de los estudiantes, sino que está diseñado para reflejar patrones plausibles de aprendizaje, fatiga y rendimiento.

**Fuente:** *https://www.kaggle.com/datasets/mabubakrsiddiq/students-learning-trajectory*

**Columnas / Artículos**
1. student_id – Identificador único para cada estudiante.
2. semana – Número de semana (1–16) del semestre.
3. horas_de_estudio – Horas dedicadas al estudio en una semana (0–10).
4. horas_de_sueño – Promedio de horas de sueño diarias (4–9), afecta la fatiga y el rendimiento.
5. stress_level – Nivel de estrés del estudiante (0–10), influye en la fatiga y las calificaciones.
6. tasa_de_asistencia – Fracción de clases a las que se asiste (0–1), influye en las tareas y el rendimiento.
7. screen_time_hours – Tiempo promedio diario frente a la pantalla para el ocio (1–10 horas).
8. ingesta de cafeína – Tazas de café/té consumidas por día (0–6).
9. learning_efficiency – Factor latente (0–1) que representa la eficacia con la que el tiempo de estudio se convierte en rendimiento; aumenta a lo largo del semestre.
10. fatigo_index – Representa el cansancio en función del estrés y el sueño; los valores más altos reducen el rendimiento.
11. Puntuación del cuestionario : puntuación semanal del cuestionario (0–100), influenciada por el estudio, la eficiencia y la fatiga.
12. assignment_score – Puntuación semanal de las tareas (0–100), influenciada por la eficiencia y la asistencia.
13. performance_index – Rendimiento semanal general, suma ponderada de las puntuaciones de los cuestionarios y las tareas.

## Hito 2: Limpieza y Preparación 🧹

En esta sección vamos a cargar los datos y realizar una auditoría inicial para entender con qué estamos trabajando. Luego, aplicaremos las técnicas de limpieza necesarias.

```python
#Carga de datos
df = pd.read_csv('data/raw/student_learning_trajectory_noisy.csv')

#Auditoría inicial
print("--- Primeras 5 filas del dataset ---")
display(df.head())

print("\n--- Información del dataset (Tipos de datos y nulos) ---")
df.info()

print("\n--- Cantidad de valores nulos por columna ---")
print(df.isnull().sum())

print(f"\nCantidad de filas duplicadas: {df.duplicated().sum()}")
```

### Aplicación de Limpieza (Data Cleaning)
A continuación, procesaremos los datos nulos, corregiremos tipos de datos (por ejemplo, si las notas se leen como texto en lugar de números) y renombraremos columnas si es necesario.

```python
import pandas as pd

df = df.copy()

#Normalizar nombres de columnas
df.columns = df.columns.str.lower().str.strip()

#Eliminar duplicados
df = df.drop_duplicates()

#Corregir tipos de datos problemáticos
df["study_hours"] = pd.to_numeric(df["study_hours"], errors="coerce")
df["student_id"] = df["student_id"].astype("Int64").astype(str)
df["week"] = df["week"].astype("Int64")

#Manejo de nulos
df = df.dropna(subset=["student_id", "week", "performance_index"])

cols_numericas = [
    "study_hours",
    "sleep_hours",
    "stress_level",
    "screen_time_hours",
    "caffeine_intake",
    "learning_efficiency",
    "fatigue_index",
    "quiz_score",
    "assignment_score"
]

for col in cols_numericas:
    if col == "assignment_score":
        df[col] = df[col].fillna(df[col].mean())
    else:
        df[col] = df[col].fillna(df[col].median())
```

```python
import numpy as np

#Correccion de valores fuera de rango
df["study_hours"] = df["study_hours"].clip(0, 12)
df["sleep_hours"] = df["sleep_hours"].clip(3, 10)
df["stress_level"] = df["stress_level"].clip(0, 10)
df["attendance_rate"] = df["attendance_rate"].clip(0, 1)
df["screen_time_hours"] = df["screen_time_hours"].clip(0, 12)
df["caffeine_intake"] = df["caffeine_intake"].clip(0, 6)
df["quiz_score"] = df["quiz_score"].clip(0, 100)
df["assignment_score"] = df["assignment_score"].clip(80, 110)

#Eliminacion de outliers con z-score
outlier_cols = [
    "study_hours",
    "sleep_hours",
    "stress_level",
    "attendance_rate",
    "screen_time_hours",
    "caffeine_intake",
    "learning_efficiency",
    "fatigue_index",
    "quiz_score",
    "assignment_score",
    "performance_index"
]

z_scores = ((df[outlier_cols] - df[outlier_cols].mean()) / df[outlier_cols].std(ddof=0)).abs()
df = df[(z_scores < 3).all(axis=1)].copy()

#Filtro minimo y orden temporal
df = df[df["attendance_rate"].between(0, 1)]
df = df.sort_values(["student_id", "week"])
df = df.reset_index(drop=True)
```

```python
df["constancy_index"] = df["study_hours"] * df["attendance_rate"]
df["real_fatigue"] = df["stress_level"] / (df["sleep_hours"] + 1)

df["risk_level"] = (
    (df["real_fatigue"] > 0.9) &
    (df["sleep_hours"] < 6.3) &
    (df["stress_level"] > 6.8)
).astype(int)

df["adjusted_performance"] = df["performance_index"] / (df["study_hours"] + 1)
df["high_screen_time"] = (df["screen_time_hours"] > 7).astype(int)

q1 = df["performance_index"].quantile(0.25)
q2 = df["performance_index"].quantile(0.50)
q3 = df["performance_index"].quantile(0.75)

def categorize_student(row):
    if row["performance_index"] >= q3:
        return "Excelente"
    elif row["performance_index"] >= q2:
        return "Alto"
    elif row["performance_index"] >= q1:
        return "Medio"
    else:
        return "Bajo"

df["student_profile"] = df.apply(categorize_student, axis=1).astype("category")
df["performance_index"] = df["performance_index"].clip(0, 110)
df = df.drop_duplicates().reset_index(drop=True)

# Validación final
print("\nNuevas columnas:\n", df.columns)
print("\nDistribución de riesgo:\n", df["risk_level"].value_counts())
print("\nPerfiles de estudiante:\n", df["student_profile"].value_counts())
print("Shape final:", df.shape)
print("\nNulos restantes:\n", df.isnull().sum())
print("\nTipos de datos:\n", df.dtypes)
print("\nResumen estadístico:\n", df.describe())

output_path = "data/processed/student_learning_trajectory_clean.csv"
df.to_csv(output_path, index=False)
print(f"\nDataset limpio guardado en: {output_path}")
```
